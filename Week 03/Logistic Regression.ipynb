# Classification and Representation

## Classification

Classification is when the answer to a problem is discrete - for example, is an email spam or not? Is a tumor cancerous?



Here is an example of a two-class classification problem trying to figure out if a tumor is malignent or benign.



We could use our previous hypothesis model, linear regression, and fit a straight line through it but that causes errors when predicting. 

Also, in classification, y is either 0 or 1. But in linear regression, our y can take on values greater than 1 or even less than zero. 

Logistic regression will output hypothesis values 0 <= h <= 1.

## Hypothesis Representation

We use the sigmoid function (sigmoid and logistic are the same). 

<center>$h_\theta(x) = g(\theta^Tx) $</center>

<center>$g(z) = \frac{1}{1+e^{-z}}$</center>



When h(x) outputs some number, we treat that number as the estimated probability that y=1 on input x. 

<center>$h_\theta(x) = P(y=1|x;\theta)$</center>

## Decision Boundary

We want to better understand when the hypothesis will make predictions as to when y=1 or when y=0. 

Suppose we predict y=1 when h(x) >= 0.5. Then y=0 when h(x) < 0.5. 

If we look at the sigmoid function, we notice that g(z) >= 0.5 when z >= 0



And since z = $\theta^Tx$, then h(x) >= 0.5, and therefore y =1, whenever $\theta^Tx$ is >= 0. The reverse is true for y=0.

This is dependent on our selection of saying y=1 whenever h(x) >= 0.5. Had we chose different starting conditions, then the results would be different. 

Suppose we have a training set like that shown here:



And $h_\theta(x) = g(\theta_0 + \theta_1x_1 + \theta_2x_2)$. Suppose we choose the thetas (through a procedure to be discussed later) $\theta_0 = -$, $\theta_1 = 1$, and $\theta_2 = 1$. 

Our theta then looks like: 

<center>$\theta = \begin{bmatrix}
-3\\
1\\
1
\end{bmatrix}$</center>

Given this. Let's try to figure out where the hypothesis will predict y=1 and y=0. 

We predict that $y=1$ if $-3 + x_1 + x_2 \geq 0$

Remeber that $-3 + x_1 + x_2 = \theta^Tx$



The decision boundary is a property of the hypothesis and its parameters and not the dataset.
