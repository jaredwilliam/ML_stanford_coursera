# Normal Equation

The normal equation gives us a much better way of solving for solving for values of the parameters in some linear regression problems. 

We can use the gradient descent algorithm, but the norml equation allows us to solve for $\theta$ analytically. 

We take the partial derivative of the cost function with respect to every $\theta$, set all of them to 0, solve each one, and you have the values of $\theta$ that minimize $J(\theta)$.

This is a pretty ugly derivation sooooo.. let's skip it. 

Let go through an example. Let's say we have four training examples, so $m=4$. 



We add in our extra column $x_0$ and create a matrix $X$. 





We also create a vector for y... 



$X$ is going to be a $m\times(n+1)$ dimensionl matrix, and $y$ is going to be an $m\times1$ vector. 

We can then calculate out $\theta$'s with the equation:

<center>$\theta = (X^TX)^{-1}X^Ty$</center>

If using the normal equation, then feature scaling isn't needed. 

Gradient Descent vs Normal Equation:
- Gradient Descent:
    - Need to choose $\alpha$
    - Need many iterations
    - Works wel even when $n$ is large
- Normal Equation:
    - No need to choose $\alpha$
    - Don't need to iterate
    - Need to compute $\theta = (X^TX)^{-1}X^Ty$ (n by n matrix)
    - Slow if $n$ is large (n > 100,000)


# Non-Invertability

If $X^TX$ is non-invetible, then you may either have redundant features (linearly dependent) or too many features (m < n)
