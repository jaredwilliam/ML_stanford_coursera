Now we are going to have multiple features to try predict the the outcome. 



So now $h_\theta(x)$ is:

<center>$h_\theta(x) = \theta_0 + \theta_1x_1 + \theta_2x_2 + ... + \theta_nx_n$</center>

We define $x_0 = 1$. This means we have a feature vector with all ones, $x^{(i)}_0=1$

Therefore, 

$x= \begin{bmatrix}
           x_{0} \\
           x_{1} \\
           x_{2} \\
           \vdots \\
           x_{n}
         \end{bmatrix}$

$\theta = \begin{bmatrix}
           \theta_{0} \\
           \theta_{1} \\
           \theta_{2} \\
           \vdots \\
           \theta_{n}
         \end{bmatrix}$
     
$h_\theta(x) = \theta^Tx$

# Gradient Descent for Multiple Variables

<center>$J(\theta_0,\theta_1,\cdots,\theta_n) = J(\theta) = \frac{1}{2m} \sum_{i=1}^m [h_\theta(x^{(i)}) - y^{(i)}]^2$</center>

Repeat until convergence:

<center>$\theta_j := \theta_j-\alpha \frac{\partial}{\partial \theta_j} J(\theta_0,\theta_1,\cdots,\theta_n)$</center>

simultaneously update for every $j=0,\cdots,n$

$\theta_j := \theta_j-\alpha \frac{1}{m}\sum_{i=1}^m [h_\theta(x^{(i)}) - y^{(i)}]x^{(i)}_j$

# Feature Scaling

Make sure features are on a similar scale. This allows the contours to be much more circular and allows gradient descent to run faster. 

We want to generally get every feature into approximately a -1 to +1 rage. Really anything on a scale of -3 to +3 and -1/3 to +1/3 should be fine. 

**Mean Normalization**

Replace $x_i$ with $x_i - \mu_i$ to make features have approximately zero mean (Do not apply to $x_0=1$)

For example, say that a feature is the number of bedrooms in a home. The range of bedrooms is 1-5 and the average is 2. We can scale x by:

<center>$x = \frac{no.\;bedrooms-2}{5}$</center>

Can also divide by the range, or max value, or standard deviation. As long as the final result gets the featues into the approprate range for gradient descent. 

# Learning Rate

<center>$\theta_j := \theta_j-\alpha \frac{\partial}{\partial \theta_j} J(\theta)$</center>

We are going to discuss 
- "debugging" which is how to make sure gradient descent is working 
- How to choose a learning rate $\alpha$

To make sure gradient descent is working, you can plot the cost function values vs the number of iterations of the algorithm that have run. You want a curve that continually goes to a smaller value



We can also have an automatic convergence test. For example, declare convergence if $J(\theta)$ decreases by less than $10^{-3}$ in one iteration. That threshold can be difficult to select though. 

If J is increasing, then you likely need a smaller learning rate. 

Selecting a learning rate is really sort of a guess and check process. 0.001, 0.003, 0.01, 0.03,...
